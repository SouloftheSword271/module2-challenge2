```python
#import required packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
#get movie title from webpage
def get_title(soup):
    title = soup.find('div', class_='category-page__members')
    title_list = [li.find('a').attrs['title'] for li in title.findAll('li')]
    return title_list
```

```python
#get movie url from webpage
def get_url(soup):
    url = soup.find('div', class_='category-page__members')
    url_list = [li.find('a').attrs['href'] for li in url.findAll('li')]
    return url_list
```

```python
#used to navigate through the film list pages
def get_next_page(soup, titles):      
    try:
        if titles[-1] == 'Cannibal Holocaust':
            next_button = soup.find('div', class_='category-page__pagination')
            return next_button.find('a').attrs['href']
        else:
            next_button = soup.find('a', class_='category-page__pagination-next wds-button wds-is-secondary')
            return next_button.attrs['href']
    except:
        return None
```

```python
#function used in tandem with the previous 3
#to collect movie titles and urls
def parse_url(url, titles, urls):
    html_page = requests.get(url)
    soup = BeautifulSoup(html_page.content, 'lxml')
    titles += get_title(soup)
    urls += get_url(soup)
    time.sleep(2)
    print(titles)
    next_url = get_next_page(soup, titles)
    time.sleep(2)
    print(urls)
    if next_url:
        return parse_url(next_url, titles, urls)
    else:
        urls = ['https://horror.fandom.com/' + i for i in urls]
        return titles, urls
```

```python
#function used for obtaining information from the infobox
def get_info(urls, tag, feature):
    info = []
    for i in urls:
        html_page = requests.get(i)
        soup = BeautifulSoup(html_page.content, 'lxml')
        try:
            feat = soup.find(tag, text=feature).find_next(tag).text.strip()
            info.append(feat)
            time.sleep(2)
            print(feat)
        except:
            info.append(' ')
    return info
```

```python
#function to get deaths since they're formatted differently than the infobox
def get_deaths(urls):
    deaths = []
    for i in urls:
        html_page = requests.get(i)
        soup = BeautifulSoup(html_page.content, 'lxml')
        try:
            death = len(pd.read_html(str(soup.find('table', attrs={"class":'mw-collapsible'})))[0])
            deaths.append(death)
            time.sleep(2)
            print(death)
        except:
            deaths.append(' ')
    return deaths
```

```python
#variable declaration
url = "https://horror.fandom.com/wiki/Category:Films"
titles = []
urls = []
#call parse_url to begin operation
titles, urls = parse_url(url, titles, urls)
```

```python
#variable declaration
dates = []
gross = []
runtime = []
deaths = []
```

```python
#call get_deaths
deaths = get_deaths(urls)
```

```python
#get gross, if available
gross = get_info(urls, 'div', 'Gross')
```

```python
#get runtime, if available
runtime = get_info(urls, 'div', 'Runtime')
```

```python
#get release date, if available
dates = get_info(urls, 'div', 'Release Date(s)')
```

```python
#create dataframe with lists and appropriate column titles
df = pd.DataFrame([titles, urls, runtime, dates, gross, deaths]).transpose()
df.columns = ['Title', 'URL', 'Runtime', 'Release', 'Gross', 'Deaths']
print(len(df))
df.head()
```
